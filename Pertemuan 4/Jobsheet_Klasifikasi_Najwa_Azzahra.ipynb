1. `from sklearn.svm import SVC`: SVC adalah model klasifikasi berbasis Support Vector Machine (SVM), yang digunakan untuk mengklasifikasikan data dengan menemukan garis atau hyperplane yang memisahkan kelas-kelas data secara optimal.

    - `Kernel `adalah fungsi yang menentukan bagaimana SVM memisahkan data.
    - `'linear' `berarti kita ingin menggunakan linear kernel, yang berarti model akan mencoba menemukan garis lurus (atau hyperplane dalam dimensi lebih tinggi) untuk memisahkan kelas-kelas data.
4. `C=1E10`:
    - `C` adalah parameter regulasi. Semakin besar nilai C, semakin kecil penalti untuk kesalahan klasifikasi.
    - `1E10 (1  10^10)` adalah nilai yang sangat besar, sehingga model akan memprioritaskan menemukan garis yang memisahkan data dengan sangat baik (overfitting mungkin terjadi jika C terlalu besar, karena model tidak akan mentolerir kesalahan sedikitpun).

    - `fit(X, y)` adalah langkah untuk melatih model.

    - Pada baris ini, model SVM dilatih menggunakan data X dan y untuk menemukan hyperplane terbaik yang memisahkan kelas-kelas dalam data.

````python
# Import library yang diperlukan
from sklearn.svm import SVC
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt

# Membuat data
X, y = make_blobs(n_samples=50, centers=2, random_state=0, cluster_std=0.60)

# Membuat dan melatih model SVC (Support Vector Classifier)
model = SVC(kernel='linear', C=1E10)
model.fit(X, y)

# Fungsi untuk plot decision boundary dan margin
def plot_svc_decision_function(model, ax=None, plot_support=True):

    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    # Buat grid untuk evaluasi model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)

    # Plot batas dan margin
    ax.contour(X, Y, P, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])

    # Plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
                   s=300, linewidth=1, facecolors='none')
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)

# Plot scatter plot dari data
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')

# Memanggil fungsi untuk menampilkan decision boundary SVM
plot_svc_decision_function(model)

# Menampilkan plot
plt.show()
````

**Analisis** Kode tersebut digunakan untuk membuat dan melatih model klasifikasi menggunakan Support Vector Classifier (SVC), kemudian memvisualisasikan decision boundary (batas keputusan) dari model tersebut beserta data yang sudah dilatih.

Namun, garis ini tidak efektif dalam memisahkan kedua kelompok data berbentuk lingkaran konsentris.

Setiap kali kode dijalankan hasilnya akan berbeda-beda karena adanya noise (gangguan acak) dalam data yang dihasilkan oleh fungsi make_circles dan tidak adanya random_state

1. `X, y = make_blobs(n_samples=100, centers=2, random_state=0, cluster_std=1.2)`:
    - `make_blobs()`: Fungsi dari sklearn.datasets yang digunakan untuk menghasilkan dataset buatan berupa titik-titik (blobs) yang dikelompokkan ke dalam beberapa klaster.
    - `n_samples=100`: Dataset terdiri dari 100 sampel atau titik.
    - `centers=2`: Dataset akan memiliki 2 pusat klaster (kelompok).
    - `random_state=0`: Menjamin bahwa dataset yang dihasilkan selalu sama setiap kali kode ini dijalankan.
    - `cluster_std=1.2`: Menentukan standar deviasi dari setiap klaster, Nilai standar deviasi yang lebih besar (seperti 1.2) membuat titik-titik tersebar lebih luas, sehingga lebih mungkin terjadi overlap antara klaster yang berbeda.
    - `X`: Berisi koordinat titik-titik data dalam ruang dua dimensi (fitur x dan y).
    - `y`: Berisi label (0 atau 1) yang mengindikasikan klaster dari setiap titik dalam dataset.

2. `plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn');`:
    - `plt.scatter()`: Fungsi dari matplotlib.pyplot untuk membuat scatter plot, yaitu plot yang menggambarkan titik-titik data.
    - `X[:, 0] dan X[:, 1]`: Memilih dua fitur (kolom) dari dataset X yang akan diplot pada sumbu x dan y.
    - `c=y`: Memberi warna pada titik berdasarkan label y (dua kelas berbeda).
    - `s=50`: Menentukan ukuran titik-titik data menjadi 50.

````python
X, y = make_blobs(n_samples=100, centers=2,
                  random_state=0, cluster_std=0.8)

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)

for axi, C in zip(ax, [10.0, 0.1]):
    model = SVC(kernel='linear', C=C).fit(X, y)
    axi.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
    plot_svc_decision_function(model, axi)
    axi.scatter(model.support_vectors_[:, 0],
                model.support_vectors_[:, 1],
                s=300, lw=1, facecolors='none');
    axi.set_title('C = {0:.1f}'.format(C), size=14)
````

3. `fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)`: Menyesuaikan jarak antara kedua subplot agar tampilannya lebih rapi.
4. `for axi, C in zip(ax, [10.0, 0.1])`:: Melakukan iterasi untuk dua subplot. Setiap iterasi menggunakan nilai C yang berbeda (10.0 dan 0.1).
  - C adalah parameter dalam SVM yang mengontrol regularisasi.
      - C besar (10.0): Model akan lebih ketat dalam memisahkan data, dengan margin kecil dan fokus pada klasifikasi yang tepat.
      - C kecil (0.1): Model akan lebih fleksibel, dengan margin yang lebih besar, tetapi mungkin mengizinkan beberapa kesalahan klasifikasi.
      
5. model = SVC(kernel='linear', C=C).fit(X, y): Membangun dan melatih model SVM dengan kernel linear pada data X dan label y, menggunakan nilai C yang ditentukan untuk setiap subplot.

6. axi.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn'): Membuat scatter plot untuk data X, di mana warna titik (merah dan kuning) ditentukan oleh label y.

7. plot_svc_decision_function(model, axi): Menggambar decision boundary (garis pemisah) dan margin di setiap subplot sesuai dengan model SVM yang telah dilatih.

8. axi.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1], s=300, lw=1, facecolors='none'): Menampilkan support vectors di setiap subplot. Support vectors adalah titik-titik data yang mempengaruhi pembuatan decision boundary. Support vectors ditampilkan dengan ukuran lebih besar (300), memiliki outline, tetapi tidak diisi.

Fungsi ini mengunduh dataset LFW dan menyimpannya dalam variabel faces, fungsi ini menyimpan hanya orang-orang yang memiliki setidaknya 60 gambar wajah yang akan dimasukkan dalam dataset

````python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(3, 5)
for i, axi in enumerate(ax.flat):
    axi.imshow(faces.images[i], cmap='bone')
    axi.set(xticks=[], yticks=[],
            xlabel=faces.target_names[faces.target[i]])
````

````python
from sklearn.svm import SVC
from sklearn.decomposition import PCA as RandomizedPCA
from sklearn.pipeline import make_pipeline

pca = RandomizedPCA(n_components=150, whiten=True, random_state=42)
svc = SVC(kernel='rbf', class_weight='balanced')
model = make_pipeline(pca, svc)
````

````python
# pemisahan data training dan data testing

from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(faces.data, faces.target,
                                                random_state=42)
````

````python
from sklearn.model_selection import GridSearchCV
param_grid = {'svc__C': [1, 5, 10, 50],
              'svc__gamma': [0.0001, 0.0005, 0.001, 0.005]}
grid = GridSearchCV(model, param_grid)

%time grid.fit(Xtrain, ytrain)
print(grid.best_params_)
print(grid.best_score_)
````

````python
model = grid.best_estimator_
yfit = model.predict(Xtest)
````

````python
# hasil label pada data testing

fig, ax = plt.subplots(4, 6)
for i, axi in enumerate(ax.flat):
    axi.imshow(Xtest[i].reshape(62, 47), cmap='bone')
    axi.set(xticks=[], yticks=[])
    axi.set_ylabel(faces.target_names[yfit[i]].split()[-1],
                   color='black' if yfit[i] == ytest[i] else 'red')
fig.suptitle('Predicted Names; Incorrect Labels in Red', size=14)
````

````python
from sklearn.metrics import classification_report
print(classification_report(ytest, yfit,
                            target_names=faces.target_names))
````

````python
# bentuk confusion matrix

import seaborn as sns
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(ytest, yfit)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=faces.target_names,
            yticklabels=faces.target_names)
plt.xlabel('true label')
plt.ylabel('predicted label')
````

````python
def standarized_input(image):
    # resize to w: 1100, h:600
    std_img = cv2.resize(image, (1100,600))

    return std_img
````

**Analisis** standarized_input digunakan untuk mengubah ukuran gambar

````python
def label_encoder(label):
    # Encode the label
    # day as 1; night as 0
    num_val = 0

    if(label == 'day'):
        num_val = 1

    return num_val
````

**Analisis** Fungsi label_encoder digunakan untuk mengonversi kata menjadi numerik
1. num_val = 0: Variabel ini diinisialisasi sebagai 0, yang akan digunakan untuk label "night"

````python
def preprocess(img_list):
    std_img_list = []

    for item in img_list:
        image = item[0]
        label = item[1]

        # Standarized the image
        std_img = standarized_input(image)

        # Create the label
        img_label = label_encoder(label)

        std_img_list.append((std_img, img_label))

    return std_img_list
````

### Langkah 4 - Ekstraksi Fitur

````python
# Get feature based on average brightness using HSV colorspace
def avg_brightness(image):
    # Convert image to HSV
    img_hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

    # Calculate the avg of brightness
    sum_brightness = np.sum(img_hsv[:,:,2]) # take the 3rb value which is the V channel
    area = image.shape[0] * image.shape[1]
    avg = sum_brightness / area

    return avg
````

**Analisis** Fungsi avg_brightness(image) digunakan untuk menghitung rata-rata kecerahan (brightness) sebuah gambar dengan menggunakan HSV colorspace.
1. Konversi gambar ke HSV:
  - img_hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV): Fungsi ini mengubah gambar dari RGB (Red, Green, Blue) menjadi HSV (Hue = warna, Saturation = kejenuhan warna, Value=kecerahan) menggunakan OpenCV.
2. Mengambil Channel Kecerahan (V channel):
  - img_hsv[:,:,2] mengambil V channel dari gambar HSV, yang merepresentasikan kecerahan (brightness).
  - np.sum() menjumlahkan seluruh nilai piksel dari V channel untuk mendapatkan total kecerahan dari semua piksel.
3. Menghitung Area Gambar:
  - image.shape[0] dan image.shape[1] adalah dimensi tinggi dan lebar gambar.
  - area adalah total jumlah piksel dalam gambar (tinggi x lebar).
4. Menghitung Rata-rata Kecerahan
  - avg = sum_brightness / area : Rata-rata kecerahan dihitung dengan membagi total kecerahan dengan area gambar.
5. return avg:Nilai rata-rata kecerahan dikembalikan sebagai hasil dari fungsi.

### Langkah 5 - Klasifikasi dengan Metode Threshold

````python
def predict_label(img, threshold):
    # Computer average brightness
    avg = avg_brightness(img)
    pred = 0

    # Predict the label based on user defined threshold
    if avg > threshold:
        pred = 1

    return pred
````

**Analisis** Fungsi predict_label(img, threshold) digunakan untuk memprediksi label gambar berdasarkan rata-rata kecerahan (brightness) gambar, dengan menggunakan ambang batas (threshold) yang ditentukan oleh pengguna. Ini berguna untuk klasifikasi biner, seperti membedakan antara "siang" dan "malam" berdasarkan kecerahan gambar.

1. Menghitung Rata-rata Kecerahan:
Menggunakan fungsi avg_brightness(img) untuk menghitung rata-rata kecerahan dari gambar img
2. Inisialisasi prediksi: Prediksi awalnya diinisialisasi sebagai 0. Ini bisa mewakili kategori pertama, misalnya "malam"
3. Membandingkan dengan Threshold: if avg > threshold: pred = 1 Jika rata-rata kecerahan gambar avg lebih besar dari threshold, label prediksi diubah menjadi 1. Ini bisa mewakili kategori kedua, misalnya "siang".
  - threshold adalah nilai ambang batas



````python
# Test the classifier on train data
rand_img = np.random.randint(0, len(train_std_img_list))

pred = predict_label(train_std_img_list[rand_img][0], threshold=120)

# Evaluate
print(f'Image {rand_img}')
print(f'Actual label: {train_std_img_list[rand_img][1]}')
print(f'Predicted label: {pred}')
plt.imshow(train_std_img_list[rand_img][0])
````

### Langkah 6 - Evaluasi (Manual)

````python
def evaluate(img_list, threshold):
    miss_labels = []

    for file in img_list:
        # Get the ground truth / correct label
        img = file[0]
        label = file[1]

        # Get prediction
        pred_label = predict_label(img, threshold)

        # Compare ground truth and pred
        if pred_label != label:
            miss_labels.append((img, pred_label, label))

    total_img = len(img_list)
    corr_pred = total_img - len(miss_labels)
    accuracy = corr_pred / total_img

    print(f'Accuracy: {accuracy:.4f}')
````

**Analisis** Fungsi evaluate(img_list, threshold) berfungsi untuk mengevaluasi kinerja model klasifikasi gambar dengan cara membandingkan label yang diprediksi oleh model dengan label yang sebenarnya.

2.

    miss_labels.append((img, pred_label, label))
    - Membandingkan label prediksi (pred_label) dengan label sebenarnya (label).
    - Jika kedua label tersebut tidak sama, maka informasi tentang gambar, label prediksi, dan label sebenarnya ditambahkan ke daftar miss_labels.

5. total_img = len(img_list): Menghitung total gambar dalam img_list
6. corr_pred = total_img - len(miss_labels): Menghitung jumlah prediksi yang benar (corr_pred) dengan mengurangi jumlah kesalahan dari total gambar
7. accuracy = corr_pred / total_img: Menghitung akurasi dengan membagi jumlah prediksi yang benar dengan total gambar.

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from pathlib import Path
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder
````

````python
train_dir = "/content/drive/MyDrive/SEMESTER 5/MACHINE LEARNING/File/images/images/test/"
test_dir = "/content/drive/MyDrive/SEMESTER 5/MACHINE LEARNING/File/images/images/test/"
````

````python
def load_dataset(img_dir):
    p = Path(img_dir)
    dirs = p.glob('*')
    img_list = []
    for dir in dirs:
        label = str(dir).split('/')[-1]
        for file in dir.glob('*.jpg'):
            img = mpimg.imread(file)
            if img is not None:
                img_list.append((img, label))
    print(f"Memuat {len(img_list)} gambar dari {img_dir}")  # Baris debug
    return img_list
````

````python
def extract_histogram_features(image, bins=256):
    if len(image.shape) == 3:
        image = np.mean(image, axis=2)
    hist = np.histogram(image.flatten(), bins=bins, range=[0, 256])[0]
    return hist / np.sum(hist)  # Normalisasi histogram
````

````python
def prepare_data(data):
    X = np.array([extract_histogram_features(img) for img, _ in data])
    y = np.array([label for _, label in data])

    le = LabelEncoder()
    y = le.fit_transform(y)

    return X, y, le
````

````python
train_data = load_dataset(train_dir)
test_data = load_dataset(test_dir)
````

````python
X_train, y_train, le = prepare_data(train_data)
X_test, y_test, _ = prepare_data(test_data)
````

````python
print("Bentuk X_train:", X_train.shape)
print("Bentuk y_train:", y_train.shape)
````

````python
if X_train.size > 0 and y_train.size > 0:
    svm_model = SVC(kernel='rbf', C=1.0, random_state=50)
    svm_model.fit(X_train, y_train)
````

````python
y_pred = svm_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Skor Akurasi: {accuracy:.4f}")
print("Laporan Klasifikasi:")
print(classification_report(y_test, y_pred, target_names=le.classes_))
print("\nConfusion Matriks:")
print(confusion_matrix(y_test, y_pred))
````
